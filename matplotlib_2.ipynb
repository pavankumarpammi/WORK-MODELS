```python
import numpy as np
import matplotlib.pyplot as plt
x=[2,5,6,8]
y=[6,5,8,3]
plt.plot(x,y,color='g',marker="*",linewidth=5,linestyle="dashed",ms=20,markerfacecolor="r")
plt.title("GRAPH")
plt.xticks(range(1,10,2))
plt.yticks(range(1,10,2))

```

```python
overs=[0,4,8,10,15,20]
team1=[6,20,35,68,95,130]
team2=[12,25,36,45,100,150]
plt.plot(overs,team1,marker='*',label='team1')
plt.plot(overs,team2,marker='*',label='team2')
plt.title('analysis of score')
plt.xlabel("overs")
#plt.yticks(range(0,200))
plt.ylabel("score")
plt.legend()
plt.grid()
```

```python
fig=plt.figure()
x1=[2,5,9,12,15]
x2=[6,8,9,12,16]
y1=np.random.randint(0,20,5)
y2=np.random.randint(0,20,5)
ax1=fig.add_subplot(221)
ax2=fig.add_subplot(222)
ax3=fig.add_subplot(223)
ax4=fig.add_subplot(224)

ax1.plot(x1,y1)
ax2.plot(x2,y2)
ax3.plot(x1,y1)
ax4.plot(x2,y2)
ax1.grid()
ax2.grid()
ax3.grid()
ax4.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[5,3,6,9,15,4]
y=[6,8,2,4,9,15]
z=np.random.randint(0,20,6)
plt.scatter(x,z,color='red')
#plt.plot(x,z,color="red")
plt.scatter(x,y,color='green')
#plt.plot(x,z,color="green")
plt.scatter(y,z,color='orange')
#plt.plot(x,z,color="orange")
plt.xticks(range(0,20))
plt.yticks(range(0,20))

```
